**Menyiapkan fungsi preprocessing**

```python
import re
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from langdetect import detect

# Inisialisasi stemmer dan stopwords
factory = StemmerFactory()
stemmer = factory.create_stemmer()

stop_words = set([
    'yang', 'untuk', 'pada', 'dengan', 'akan', 'dan', 'di', 'ke', 'dari', 'ini',
    'itu', 'karena', 'ada', 'saya', 'kamu', 'dia', 'mereka', 'kami', 'kita',
    'tidak', 'bukan', 'atau', 'sebagai', 'juga', 'dalam', 'oleh', 'pada',
    'sudah', 'belum', 'lagi', 'agar', 'bagi', 'dapat', 'setelah', 'jadi'
])

# Filtering
def filtering(text):
    text = re.sub(r'http\S+', '', text)     # hapus URL
    text = re.sub(r'#\w+', '', text)        # hapus hashtag
    text = re.sub(r'@\w+', '', text)        # hapus mention
    text = re.sub(r'\d+', '', text)         # hapus angka
    text = re.sub(r'[^\w\s]', '', text)     # hapus simbol/punctuation
    return text

# Case Folding
def case_fold(text):
    return text.lower()

# Tokenizing
def token(text):
    return word_tokenize(text)

# Stemming
def stemming(tokens):
    return [stemmer.stem(token) for token in tokens]

# Stopword Removal
def remove_stopwords(tokens):
    return [token for token in tokens if token not in stop_words]

# Gabungan fungsi preprocessing + filter bahasa
def process_it(text):
    try:
        # Deteksi bahasa, lanjut hanya kalau Indonesia
        if detect(text) != 'id':
            return {
                'Filtering': '',
                'Case_Folding': '',
                'Tokenizing': [],
                'Stemming': [],
                'Stopwords_Removal': []
            }
    except:
        # Kalau gagal deteksi (teks kosong atau aneh)
        return {
            'Filtering': '',
            'Case_Folding': '',
            'Tokenizing': [],
            'Stemming': [],
            'Stopwords_Removal': []
        }

    filtered = filtering(text)
    lowered = case_fold(filtered)
    tokens = token(lowered)
    stemmed = stemming(tokens)
    processed = remove_stopwords(stemmed)

    return {
        'Filtering': filtered,
        'Case_Folding': lowered,
        'Tokenizing': tokens,
        'Stemming': stemmed,
        'Stopwords_Removal': processed
    }
```

**Apply fungsi preprocessing & ditambahkan pada kolom baru**

```python
# Terapkan ke DataFrame
results = df['full_text'].apply(process_it)

# Membuat kolom baru hasil preprocessing
df['Filtering'] = results.apply(lambda x: x['Filtering'])
df['Case_Folding'] = results.apply(lambda x: x['Case_Folding'])
df['Tokenizing'] = results.apply(lambda x: x['Tokenizing'])
df['Stemming'] = results.apply(lambda x: x['Stemming'])
df['Stopwords_Removal'] = results.apply(lambda x: x['Stopwords_Removal'])

# DataFrame hasil preprocessing
processed_df = df[['full_text', 'Filtering', 'Case_Folding', 'Tokenizing', 'Stemming', 'Stopwords_Removal']].copy()

# Hapus baris yang hasil Stopwords_Removal kosong → kemungkinan bukan bahasa Indonesia
processed_df = processed_df[processed_df['Stopwords_Removal'].apply(lambda x: len(x) > 0)].reset_index(drop=True)

# Tampilkan hasil
print("Hasil Preprocessing (hanya tweet berbahasa Indonesia):\n")
display(processed_df)
```

# TF-IDF

```python
import pandas as pd
import numpy as np

dm = pd.read_csv("/content/drive/MyDrive/PTT 2025/insomnia_cleaned_260_data.csv", usecols=["clean_text"])
dm.columns = ["clean_text"]
display(dm)
```

```python
# convert list formated string to list
import ast

def convert_text_list(texts):
    # Check if the input is a string before trying to convert
    if isinstance(texts, str):
        texts = ast.literal_eval(texts)
    return [text for text in texts]

dm["tekslist"] = dm["clean_text"].apply(convert_text_list)

# Accessing multiple rows using .iloc
print("Tekslist for indices 2, 3, 8, 9, 10, 13:")
print(dm["tekslist"].iloc[[2, 3, 8, 9, 10, 13]])

print("\ntype : ", type(dm["tekslist"].iloc[0]))
```

```python
def calc_TF(document):
    #perhitungan jumlah kata
    TF_dict = {}
    for term in document:
        if term in TF_dict:
            TF_dict[term] += 1
        else:
            TF_dict[term] = 1
    #perhitungan tf
    for term in TF_dict:
        TF_dict[term] = TF_dict[term] / len(document)
    return TF_dict

dm["TF_dict"] = dm['tekslist'].apply(calc_TF)

# Accessing multiple rows using .iloc
print("Tekslist for indices 2, 3, 8, 9, 10, 13:")
print(dm["TF_dict"].iloc[[2, 3, 8, 9, 10, 13]])

print("\ntype : ", type(dm["TF_dict"].iloc[0]))
```

```python
# memunculkan nilai TF
indices = [2, 3, 8, 9, 10, 13]

for index in indices:
    print(f"\nTF values for index {index}:\n")
    print('%20s' % "term", "\t", "TF\n")
    for key in dm["TF_dict"][index]:
        print('%20s' % key, "\t", dm["TF_dict"][index][key])
```

```python
def calc_DF(tfDict):
    count_DF = {}
    for document in tfDict:
        for term in document:
            if term in count_DF:
                count_DF[term] += 1
            else:
                count_DF[term] = 1
    return count_DF

DF = calc_DF(dm["TF_dict"])
```

```python
#menghitung idf
n_document = len(dm)
def calc_IDF(__n_document, __DF):
    IDF_Dict = {}
    for term in __DF:
        IDF_Dict[term] = np.log(__n_document / (__DF[term] + 1))
    return IDF_Dict

#penyimpanan kamus idf
IDF = calc_IDF(n_document, DF)
```

```python
# memunculkan nilai IDF untuk term yang ada di indices tertentu, per dokumen

indices = [2, 3, 8, 9, 10, 13]

for index in indices:
    print(f"\nIDF values for terms in document at index {index}:\n")
    print('%20s' % "term", "\t", "IDF\n")
    if index < len(dm): # Check if index is within bounds
        # Get unique terms for the current document
        terms_in_document = set(dm["tekslist"][index])
        # Print IDF for terms in the current document, sorted alphabetically
        for term in sorted(terms_in_document):
            if term in IDF:
                print('%20s' % term, "\t", IDF[term])
```

```python
#perhitungan TF-IDF
def calc_TF_IDF(TF):
    TF_IDF_Dict = {}
    for key in TF:
        TF_IDF_Dict[key] = TF[key] * IDF[key]
    return TF_IDF_Dict

#penyimpanan variabel TF-IDF
dm["TF-IDF_dict"] = dm["TF_dict"].apply(calc_TF_IDF)
```

```python
# memunculkan nilai TF-IDF
indices = [2, 3, 8, 9, 10, 13]

for index in indices:
    print(f"\nTF-IDF values for index {index}:\n")
    print('%20s' % "term", "\t", '%10s' % "TF", "\t", '%20s' % "TF-IDF\n")
    for key in dm["TF-IDF_dict"][index]:
        print('%20s' % key, "\t", dm["TF_dict"][index][key], "\t", dm["TF-IDF_dict"][index][key])
```

```python
# Membuat Matriks TF-IDF

# Mengurutkan daftar DF (document frequency) secara menurun (descending)
sorted_DF = sorted(DF.items(), key=lambda kv: kv[1], reverse=True)[:100]

# Membuat daftar kata unik berdasarkan hasil pengurutan DF
unique_term = [item[0] for item in sorted_DF]

# Fungsi untuk membuat vektor TF-IDF dari setiap dokumen
def calc_TF_IDF_Vec(tf_idf_dict):
    # Membuat vektor nol sepanjang jumlah unique_term
    TF_IDF_vector = [0.0] * len(unique_term)

    # Jika term muncul di dokumen, masukkan nilai TF-IDF-nya ke posisi yang sesuai
    for i, term in enumerate(unique_term):
        if term in tf_idf_dict:
            TF_IDF_vector[i] = tf_idf_dict[term]
    return TF_IDF_vector

# Terapkan fungsi di atas ke setiap dokumen
dm["TF_IDF_Vec"] = dm["TF-IDF_dict"].apply(calc_TF_IDF_Vec) # Corrected column name

# Menampilkan hasil
print("Tampil baris pertama dari matriks TF-IDF Series:\n")
print(dm["TF_IDF_Vec"][0])

print("Ukuran matriks TF-IDF:", len(dm["TF_IDF_Vec"][0]))
```

```python
# Menampilkan Top 100 Term Berdasarkan Total TF-IDF

# Konversi daftar vektor TF-IDF menjadi array numpy
TF_IDF_Vec_List = np.array(dm["TF_IDF_Vec"].to_list())

# Menjumlahkan nilai TF-IDF untuk setiap kata di semua dokumen
sums = TF_IDF_Vec_List.sum(axis=0)

# Menggabungkan setiap kata dengan total TF-IDF-nya
data = []
for col, term in enumerate(unique_term):
    data.append((term, sums[col]))

#  Membuat DataFrame peringkat kata
ranking = pd.DataFrame(data, columns=['term', 'rank'])

# Mengurutkan dari yang terbesar
ranking = ranking.sort_values('rank', ascending=False)

# Menampilkan hasil (menggunakan display() untuk menampilkan seluruh DataFrame)
print("Ranking 100 Top Terms berdasarkan Total TF-IDF:\n")
display(ranking)
```
